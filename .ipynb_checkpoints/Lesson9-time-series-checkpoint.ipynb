Проверим на стационарность по KPSS и Дики-Фуллера. 

```python
# KPSS по заражениям за день
stat, p_value, *_ = statsmodels.tsa.stattools.kpss(col_d['Заражений за день'])
stat, p_value
```

```python
# KPSS по выздоровлениям за день
stat, p_value, *_ = statsmodels.tsa.stattools.kpss(col_d['Выздоровлений за день'])
stat, p_value
```

```python
# KPSS по смертям за день
stat, p_value, *_ = statsmodels.tsa.stattools.kpss(col_d['Смертей за день'])
stat, p_value
```

Вывод: так как p-value по всем колонкам менее 0,05 - ряды не стационарны

Смотрим также по тесту Дики-Фуллера (нулевая гипотеза - ряд нестационарен).

```python
# тест Дики-Фуллера по заражениям за день
stat, pvalue, *_ = statsmodels.tsa.stattools.adfuller(col_d['Заражений за день'])
stat, pvalue
```

```python
# тест Дики-Фуллера по выздоровлениям за день
stat, pvalue, *_ = statsmodels.tsa.stattools.adfuller(col_d['Выздоровлений за день'])
stat, pvalue
```

```python
# тест Дики-Фуллера по смертям за день
stat, pvalue, *_ = statsmodels.tsa.stattools.adfuller(col_d['Смертей за день'])
stat, pvalue
```

Вывод: p-value по всем показателям превышает 0.05, все ряды не стационарны.

Приведем ряды к стационарному виду

Для этого применим дифференцирование:

```python
# Функция для дифференцирования:
def stacryad(f):
    df_diff = pd.DataFrame(f.iloc[0:,0])
    for i in range(1, 4):
            df_diff[f.columns[i]] = f.iloc[:, i].diff(1)
    df_diff = df_diff.fillna(0.0)
    return df_diff
```

```python
# показатели с дифференцированием:
df_diff = stacryad(col_d)
```

```python
df_diff.head()
```

Проверка на стационарность после дифференцирования (по KPSS, Дики-Фуллер) дает такие результаты:

```python
# KPSS по заражениям за день
stat, p_value, *_ = statsmodels.tsa.stattools.kpss(df_diff['Заражений за день'])
stat, round(p_value, 4)
```

```python
# KPSS по выздоровлениям за день
stat, p_value, *_ = statsmodels.tsa.stattools.kpss(df_diff['Выздоровлений за день'])
stat, round(p_value, 4)
```

```python
# KPSS по смертям за день
stat, p_value, *_ = statsmodels.tsa.stattools.kpss(df_diff['Смертей за день'])
stat, round(p_value, 4)
```

Видим, что по KPSS ряды стали стационарными

```python
# тест Дики-Фуллера по заражениям за день
stat, pvalue, *_ = statsmodels.tsa.stattools.adfuller(df_diff['Заражений за день'])
stat, round(pvalue, 4)
```

```python
# тест Дики-Фуллера по выздоровлениям за день
stat, pvalue, *_ = statsmodels.tsa.stattools.adfuller(df_diff['Выздоровлений за день'])
stat, round(pvalue, 4)
```

```python
# тест Дики-Фуллера по смертям за день
stat, pvalue, *_ = statsmodels.tsa.stattools.adfuller(df_diff['Смертей за день'])
stat, round(pvalue, 4)
```

Если на базе продифференцированных данных (df_diff):

```python
X = df_diff['Заражений за день'].values
x_train, x_test = train_test_split(df_diff['Заражений за день'].values, shuffle = False, test_size = 31)  # test_size = 31, так как прогноз на месяц
```

```python
# Подберем лучшие параметры  с помощью ARIMA:
```

```python
def arimafu(p, q, d):
    model = None
    try:
        model = ARIMA(x_train, order = (p, q, d)).fit()
    except:
        pass
    return model

aic_min = 10000
mfit = None
pf = 0
qp = 0
dp = 0
for p in range(0, 5):
        for q in range(0, 5):
            for d in range(0, 5):
                m = arimafu(p, q, d)
                if(m == None):
                    continue
                if (m.aic < aic_min):
                    print(m.aic)
                    aic_min = m.aic
                    mfit = m
                    pf = p
                    qp = q
                    dp = d
                    print(pf, qp, dp)
```

Для ARIMA выбираем лучший вариант (c минимальным aic):

```python
model = ARIMA(x_train, order=(2, 0, 4)).fit()
forecast_t = model.forecast(31)[0]   #прогноз на месяц
y = df_diff['Заражений за день'].values
x = range(len(y))
yf = forecast_t
xf = range(len(x_train), len(x_train) + len(x_test))
```

```python
# Смотрим, как выглядит наш прогноз:
plt.plot(x, y)
plt.plot(xf, forecast_t)
```

```python
mean_absolute_error(x_test, forecast_t)
```

```python
mean_squared_error(x_test, forecast_t)
```

```python
r2_score(x_test, forecast_t)
```

Вывод: прогноз получился неважный (малореалистичный), да и на графике это видно

```python
# Обратное преобразование с помощью np.cumsum
```

```python
xf[0]
```

```python
y_sum = np.cumsum(y)
f_sum = np.cumsum(forecast_t)
f_sum = f_sum + np.cumsum(x_test)
plt.plot(x, y_sum)
plt.plot(xf, f_sum + y_sum[xf[0]])
```

```python
mean_absolute_error(np.cumsum(x_test), np.cumsum(forecast_t))
```

```python
mean_squared_error(np.cumsum(x_test), np.cumsum(forecast_t))
```

```python
r2_score(np.cumsum(x_test), np.cumsum(forecast_t))
```

Градиентный бустинг (для данных без дифференцирования):

```python
pred = []
base_pred = []
test = []
x = col_d['Смертей за день'].values

for i in range(334, len(x)):  
    y_train = x[:i]
    x_train = np.arange(len(y_train)).reshape(-1, 1)
    test.append(x[i])
    model = GradientBoostingRegressor()
    model.fit(x_train, y_train)
    forecast_t = model.predict([[i]])
    pred.append(forecast_t) 
    base_pred.append(x[i-1])
```

```python
plt.plot(test)
plt.plot(pred)
plt.plot(base_pred)
plt.legend(['test', 'pred', 'base_pred'])
```

```python
mean_absolute_error(test, pred), mean_absolute_error(test, base_pred)  
```

```python
mean_squared_error(test, pred), mean_squared_error(test, base_pred)
```

```python
r2_score(test, pred), r2_score(test, base_pred)  
```

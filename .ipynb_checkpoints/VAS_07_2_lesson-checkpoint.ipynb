```python
import numpy as np
import pandas as pd
import scipy

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score, confusion_matrix, r2_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

Оценим зависимость цен от года выпуска автомобиля

```python
vw = pd.read_csv('../data/vw.csv')
vw.sort_values('year')
vw.transmission = pd.Categorical(vw.transmission)
vw.model = pd.Categorical(vw.model)
vw
```

Выведим график зависимости года и цены

```python
sns.pairplot(vw)
```

Сконвертируем категории model и transmission в числовые коды, а fuelType в OneHotEncoding

```python
model_code = vw.model.cat.codes
trans_code = vw.transmission.cat.codes
fuel_code = pd.get_dummies(vw.fuelType)
df = pd.DataFrame()
df['model_code'] = model_code
df['trans_code'] = trans_code
df['year'] = vw.year
df['mileage'] = vw.mileage
df['tax'] = vw.tax
df['mpg'] = vw.mpg
df['engineSize'] = vw.engineSize
```

```python

df = df.join(fuel_code)
df
```

Теперь создадим numpy матрицу факторов X, и результирующий вектор Y с ценой из таблицы.
Создадим тренировочный и тестовый наборы данных

```python
x = df.values
y = vw.price.values
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

Возьмем модель линейной регресси и попробуем сделать прогноз на параметрах по умолчанию.
В качестве метрики эффективности прогноза воспользуемся r2_score (R^2)

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
r2_score(y_test, y_pred)
```

Качество прогноза составляет 78%
Попробуем построить прогноз используя случайные леса

```python
#Ставить значения по умолчанию или большую размерность нельзя - вылетает с ошибками памяти
rf = RandomForestClassifier(n_estimators=10, max_depth=10)
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
r2_score(y_test, y_pred)
```

Модель случайного леса существенно точней. Она объясняет 89% ценообразования в зависимости от характеристик автомобиля, включая его модель, год выпуска, пробег и тип двигателя

Попробуем кроссвалидацию с линейной регрессией

```python
scores = cross_val_score(lr, x_train, y_train, cv=5)
print("mean: ", scores.mean(), " std:", scores.std())
```

Попробуем кроссвалидацию со случайными лесами

```python
scores = cross_val_score(rf, x_train, y_train, cv=5)
print("mean: ", scores.mean(), " std:", scores.std())
```

Крос-валидация для случайного леса не задалась.

Наибольшим весом обладает 4 параметр таблицы: пробег (0.32). Следом по значимости с большим отрывом идут параметры год и марка автомобиля.
Учитывая что пробег прямопропорционально зависит от года автомобиля, можно сделать вывод что эти переменные связаны. Так как именно пробег является определяющим, то именно год зависит от пробега, а не пробег от года.

Подберем гиперпараметры для линейной модели и модели случайных лесов. Начнем со случайных лесов.

```python
params = {
    'n_estimators': [5, 100, 500, 1000],
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 10]
}
```

```python
search = RandomizedSearchCV(RandomForestClassifier(), params)
search.fit(x_train, y_train)
search.best_estimator_
```
